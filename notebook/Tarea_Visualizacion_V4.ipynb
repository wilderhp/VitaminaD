```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly
import plotly.graph_objs as go
import plotly.express as px
import ipywidgets as widgets
from plotly.subplots import make_subplots

import warnings
warnings.filterwarnings('ignore')

#data = pd.read_csv("DATOS ADD_VISUALIZACION.CSV", decimal=',')
data = pd.read_csv("https://raw.githubusercontent.com/wilderhp/VitaminaD/main/Vitamina_D_Visualizacion_26092022.csv",sep=';', decimal=',')
data
```

```python
data.columns = data.columns.str.upper()
data.head(5)
```

```python
# Para definir los resultados de la vitamina en base a la tabla en https://tucuerpomehabla.com/la-vitamina-d-clave-en-nuestro-bienestar/
def condicion(x):
    x=int(x)
    if x<30:
        return "Deficiente"
    else:
        return "Deseable"
    

data['RESULTADO_VAL']= 1

data['VIT_D_TXT'] = data['VITAMINA D'].apply(condicion)
data
```

```python
def agregar_proporcion(grupo):    
    grupo["PROP"] = grupo['VITAMINA D'] / grupo['VITAMINA D'].sum()
    return grupo

def orden_zona(x):
    x=str(x)
    if x=="NORTE":
        return 1
    elif x=="CENTRO":
        return 2
    else:
        return 3

data['ORDEN_ZONA'] = data['MACROZONA'].apply(orden_zona)

data.sort_values(by=['ORDEN_ZONA'], inplace=True)

data = data.reset_index()

df_filter = data.groupby(['AÑO','LABORATORIO','ADULTO_NIÑO']).apply(agregar_proporcion)

df_filter
```

```python
from numpy import average

x=df_filter[['AÑO','LABORATORIO','VITAMINA D']]

nombres = {
   ' SANTIAGO': 'Región Metropolitana de Santiago',
   ' LA SERENA':'Región de Coquimbo',
   ' VALDIVIA': 'Región de Los Ríos',
   ' ANTOFAGASTA': 'Región de Antofagasta',
   ' PUERTO MONTT': 'Región de Los Lagos',
   ' ILLAPEL':'Región de Coquimbo',
   ' RANCAGUA':"Región del Libertador General Bernardo O'Higgins",
   ' CONCEPCION': 'Región del Biobío',
   ' OVALLE':'Región de Coquimbo',
   ' LOS ANDES': 'Región de Valparaíso',
   ' VIÑA DEL MAR':'Región de Valparaíso',
   ' VALLENAR':'Región de Atacama',
   ' TEMUCO':'Región de La Araucanía',
   ' COPIAPO':'Región de Atacama',
   ' ARICA':'Región de Arica y Parinacota',
   ' CALAMA':'Región de Antofagasta', 
   ' IQUIQUE':'Región de Tarapacá',
   'TALCA': 'Región del Maule'
}
codigos = {
   ' SANTIAGO': 13,
   ' LA SERENA':4,
   ' VALDIVIA': 14,
   ' ANTOFAGASTA': 2,
   ' PUERTO MONTT': 10,
   ' ILLAPEL':4,
   ' RANCAGUA':6,
   ' CONCEPCION': 8,
   ' OVALLE':4,
   ' LOS ANDES': 5,
   ' VIÑA DEL MAR':5,
   ' VALLENAR':3,
   ' TEMUCO':9,
   ' COPIAPO':3,
   ' ARICA':0,
   ' CALAMA':2, 
   ' IQUIQUE':1,
   'TALCA': 7
}
data_ = x
data_['nom_reg'] = data_.LABORATORIO.map(lambda x: nombres.get(x))
data_['cod_reg'] = data_.LABORATORIO.map(lambda x: codigos.get(x))

tabla = data_.pivot_table("VITAMINA D", index="AÑO",
                           columns = "cod_reg", aggfunc=average).reset_index()
import plotly.graph_objects as go

#extract color palette, the palette can be changed
pal = list(sns.color_palette(palette='viridis', n_colors=len(codigos)).as_hex())


fig = go.Figure()
for d,p in zip(tabla, pal):
    fig.add_trace(go.Scatter(x = tabla[tabla.columns[1:]==d],
                             y = tabla[tabla.columns[:1]==d],
                             name = d,
                             line_color = p, 
                             fill=None))   #tozeroy 

fig.show()
```

```python
def update(year,zona):
  global df_filter
  
  fig= make_subplots(
    rows=2, cols=1,
    row_heights = [0.3, 0.7],
    specs=[[ {"type": "domain"}],
           [ {"type": "domain"}]],
    subplot_titles=(" ", "Gráfico Sunbrst \n"),
    shared_xaxes=True,
                    vertical_spacing=0.05)
 
  if zona !="ALL":
    df = df_filter.query(f"MACROZONA == '{zona.upper()}'")
  else:
    df = df_filter
    
  if year !="ALL":
    df = df.query(f"AÑO == {year}")

  # Treemap
  fig.add_trace(
    px.treemap(df, values="PROP",path = [px.Constant("Gráfico Treemap"),'MACROZONA','ESTACIONES_AÑO','VIT_D_TXT'],
                color='ESTACIONES_AÑO', hover_data=['VIT_D_TXT']
                  ).data[0],
      row=1, col=1    
    )
  #fig.update_traces(root_color="lightgrey")

  cm = {
    "NORTE": "red",
    "CENTRO": "yellow",
    "SUR": "white"
  }
  fig.for_each_trace(
    lambda t: t.update(
        marker_colors=[
            cm[id.split("/")[1]] if len(id.split("/")) == 2 else c
            for c, id in zip(t.marker.colors, t.ids)
        ]
    )
)
  fig.update_layout(margin = dict(t=25, l=25, r=25, b=25))

  # Sunburts
  fig2 = px.sunburst(df, path=['AÑO',"VIT_D_TXT", 'ADULTO_NIÑO'], values='RESULTADO_VAL')

  fig.add_trace(
      go.Figure(go.Sunburst(
        labels=fig2['data'][0]['labels'],
        parents=fig2['data'][0]['parents'],
        values=fig2['data'][0]['values'],
        ids=fig2['data'][0]['ids'],
        branchvalues='total',
        marker=dict(colorscale='YlOrRd'),
        hovertemplate='<b>%{label}:</b> <br> Total : %{value} <extra></extra>',
        maxdepth=3
      )).data[0], row=2, col=1)
  
  #fig2.update_traces(textinfo="label+percent parent")
  #names = {'Plot 1':'2016', 'Plot 2':'2017'}

  #fig.for_each_annotation(lambda a: a.update(text = a.text + ': ' + names[a.text]))
  fig.update_layout(   
    #template="plotly_white",
    #title_text="Customizing Subplot Axes",
    #margin=dict(r=0, t=0, b=0, l=0),
    height = 900,
    #width = 1200,      
  )

  fig.update_traces(textinfo="label+percent root") 
 
  fig.show()
```

```python
class SelectMultipleInteract(widgets.HBox):
    def __init__(self):
        self.W1 = widgets.Dropdown(
            value='ALL', 
            options=['ALL','2018', '2019', '2020', '2021'], 
            description='Año :'
        )


        self.W2 = widgets.Dropdown(
            value='ALL', 
            options=['ALL','Norte', 'Centro', 'Sur'], 
            description='Zona :'
        )

        self.selectors = [self.W1, self.W2]

        super().__init__(children=self.selectors)
        self._set_observes()

    def _set_observes(self):       
        for widg in self.selectors:
            widg.observe(self._observed_function, names='value')

    def _observed_function(self, widg):              
        x=[]
        SelectMultipleInteract()
        for widg in self.selectors:
            x.append(widg.get_interact_value())
       
        SelectMultipleInteract()
        update(x[0],x[1])
           
SelectMultipleInteract()
```
