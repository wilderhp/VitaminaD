```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
chile_pol = gpd.read_file('../shape_chile/chile_pol.shp')
```

```python
chile_pol['select'] = chile_pol.codreg.map(lambda x: 0 if (int(x)==8 or int(x)==16) else int(x))
```

```python
chile_pol = chile_pol.dissolve(by='select')
```

```python
plt.figure(figsize=(40,20))
chile_pol.plot()
```
