```python
temp2018 = pd.DataFrame(columns = ['CodigoNacional','NombreEstacion', 'Latitud','Altura', 'Año', 'Mes', 'Dia','T.Maxima','T.Minima'
])

for Mes in range(1,13):
    path = f"Temperaturas/temperatura_{Mes}2018.csv"
    frame = pd.read_csv(path, sep=";", low_memory =False)
    
    temp2018 = pd.concat([temp2018, frame])
```

```python
temp2019 = pd.DataFrame(columns = ['CodigoNacional','NombreEstacion', 'Latitud','Altura', 'Año', 'Mes', 'Dia','T.Maxima','T.Minima'
])

for Mes in range(1,13):
    path = f"Temperaturas/temperatura_{Mes}2019.csv"
    frame = pd.read_csv(path, sep=";", low_memory =False)
    
    temp2019 = pd.concat([temp2019, frame])
```

```python
temp2019
```

```python
temp2020 = pd.DataFrame(columns = ['CodigoNacional','NombreEstacion', 'Latitud','Altura', 'Año', 'Mes', 'Dia','T.Maxima','T.Minima'
])

for Mes in range(1,13):
    path = f"Temperaturas/temperatura_{Mes}2020.csv"
    frame = pd.read_csv(path,  sep=";", encoding='latin', low_memory =False)
    
    temp2020 = pd.concat([temp2020, frame])
```

```python
temp2021 = pd.DataFrame(columns = ['CodigoNacional','NombreEstacion', 'Latitud','Altura', 'Año', 'Mes', 'Dia','T.Maxima','T.Minima'
])

for Mes in range(1,7):
    path = f"Temperaturas/temperatura_{Mes}2021.csv"
    frame = pd.read_csv(path, sep=";", low_memory =False)
    
    temp2021 = pd.concat([temp2021, frame])
```

```python
Temp2018_2021 = pd.concat([temp2018, temp2019, temp2020,temp2021], axis=0)
```
